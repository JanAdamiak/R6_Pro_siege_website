```python
df2 = df[df['round number'] == 1]
df2 = df2[["Operator on ATK1", "Operator on ATK2", "Operator on ATK3", "Operator on ATK4", "Operator on ATK5", "Operator on ATK6", "Operator on ATK7", "Operator on ATK8", "Operator on ATK9", "Operator on ATK10", "Map name"]]
df3 = df[df['round number'] == 1]
df3 = df3[["Operator on DEF1", "Operator on DEF2", "Operator on DEF3", "Operator on DEF4", "Operator on DEF5", "Operator on DEF6", "Operator on DEF7", "Operator on DEF8", "Operator on DEF9", "Operator on DEF10", "Map name"]]
bans = df[df['round number'] == 1]
bans = bans[["1st ban", "2nd ban", "3rd ban", "4th ban", "Map name"]]
```

```python
stackeddf2 = df2.set_index('Map name').stack()
stackeddf3 = df3.set_index('Map name').stack()
stackedbans = bans.set_index('Map name').stack()
```

```python
ATK_picks = stackeddf2.reset_index()
ATK_picks.drop('level_1', axis=1, inplace=True)
DEF_picks = stackeddf3.reset_index()
DEF_picks.drop('level_1', axis=1, inplace=True)
ban_picks = stackedbans.reset_index()
ban_picks.drop('level_1', axis=1, inplace=True)
```
